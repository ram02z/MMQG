```python
from datasets import load_dataset

pmc = (
  load_dataset("pmc/open_access", split="train", streaming=True)
)

pmc_iter = iter(pmc)
```

```python
import csv

def append_row_to_csv(filename, row):
  with open(filename, 'a', newline='') as csvfile:
    fieldnames = ['accession_id', 'text']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writerow(row)

# run this function first, to create the csv file before you start appending data
def create_empty_csv(filename):
    with open(filename, 'w', newline='') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=['accession_id', 'text'])
        writer.writeheader()
```

```python

csv_filename = './../data/pmc_1mil.csv'

for idx, row in enumerate(pmc_iter):
  if idx == 1000000:
    break

  row['text'] = row['text'].replace('\n', '')
  data = {'accession_id': row['accession_id'], 'text': row['text']}
  
  append_row_to_csv(csv_filename, data)
  print(idx)
```
